```python
from snappy import jpy, ProductIO, GPF, HashMap
import os
import glob
```

```python
input_dir = "./data/"
input_files = sorted(glob.glob(os.path.join(input_dir, '*.tif')))
print(input_files)
```

```python
for input_file in input_files:

    x = 0
    y = 0
    width = 512
    height = 512
    filename = input_file.replace("./data\\", "")
    filename_tif = filename.replace(".tif", "")
    
    product = ProductIO.readProduct(input_file)

    while (True):

        SubsetOp = jpy.get_type('org.esa.snap.core.gpf.common.SubsetOp')

        HashMap = jpy.get_type('java.util.HashMap')
        GPF.getDefaultInstance().getOperatorSpiRegistry().loadOperatorSpis()
        parameters = HashMap()
        parameters.put('copyMetadata', True)
        parameters.put('region', "%s,%s,%s,%s" % (x, y, width, height))
        product_subset = GPF.createProduct('Subset', parameters, product)

        width = product_subset.getSceneRasterWidth()
        print("Width: {} px".format(width))
        height = product_subset.getSceneRasterHeight()
        print("Height: {} px".format(height))
        band_names = product_subset.getBandNames()
        print("Band names: {}".format(", ".join(band_names)))

        # ProductIO.writeProduct(product_subset, f"data/final_subset_{string}", 'BEAM-DIMAP')

        ProductIO.writeProduct(product_subset, f"data_new/final_mask_{filename_tif}_{x}_{y}", 'GeoTIFF')
    
        if (x + width == 4000):
            x = 0
            if (y + height == 3008):
                break
            elif (y + height > 3008):
                y = 2496
            else: 
                y = y+height
        elif (x + width > 4000):
            x = 3488
        else:
            x = x+width
```
